```python
import pandas as pd
from pykrx import stock
from sqlalchemy import create_engine, text
from blsh.common.env import DB_URL
from blsh.database.model import Base, Ohlcv

engine = create_engine(DB_URL, pool_recycle=600)


def save_ohlcv(date):
    df = stock.get_market_ohlcv(date, market="ALL").reset_index()
    df.columns = ['ticker', 'open', 'high', 'low', 'close', 'vol', 'val', 'change', 'cap']
    df['date'] = date
    df.to_sql(Ohlcv.__tablename__, con=engine, if_exists='append', index=False)

def get_ohlcv(date):
    with engine.connect() as conn:
        return pd.read_sql_query(text(f"SELECT * FROM  {Ohlcv.__tablename__} WHERE DATE='{date}';"), conn)    
        
     
```

```python
# 테이블 생성
Base.metadata.create_all(bind=engine)
```

```python
# 데이터 저장
save_ohlcv('20260224')
```

```python
# 데이터 조회
get_ohlcv('20260224')
```
